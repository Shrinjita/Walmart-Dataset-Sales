```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

# Assuming your data is in a CSV file named 'Walmart.csv' within the downloaded path
file_path = '/root/.cache/kagglehub/datasets/yasserh/walmart-dataset/versions/1/Walmart.csv'
df = pd.read_csv(file_path)

# Calculate the correlation between Temperature and Weekly_Sales
correlation = df['Temperature'].corr(df['Weekly_Sales'])
print(f"Correlation between Temperature and Weekly_Sales: {correlation}")

# Create a scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df)
plt.title('Temperature vs. Weekly Sales')
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle

# Assuming your data is in a CSV file named 'Walmart.csv' within the downloaded path
file_path = '/root/.cache/kagglehub/datasets/yasserh/walmart-dataset/versions/1/Walmart.csv'
df = pd.read_csv(file_path)

# Function to save a plot to a pickle file
def save_plot_to_pickle(plot, filename):
    with open(filename, 'wb') as file:
        pickle.dump(plot, file)

# Example 1: Histogram of Weekly_Sales
plt.figure(figsize=(8, 6))
hist_plot = sns.histplot(df['Weekly_Sales'], bins=30, kde=True)
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
save_plot_to_pickle(hist_plot, 'histogram.pkl')  # Save the histogram
plt.show()

# Example 2: Scatter plot of Temperature vs. Weekly_Sales
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x='Temperature', y='Weekly_Sales', data=df)
plt.title('Temperature vs. Weekly Sales')
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
save_plot_to_pickle(scatter_plot, 'scatter_plot.pkl')  # Save the scatter plot
plt.show()

# Example 3: Box plot of Weekly_Sales by Store
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x='Store', y='Weekly_Sales', data=df)
plt.title('Weekly Sales by Store')
plt.xlabel('Store')
plt.ylabel('Weekly Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
save_plot_to_pickle(box_plot, 'box_plot.pkl')  # Save the box plot
plt.show()

# Example 4: Correlation heatmap
plt.figure(figsize=(10, 8))
# Excluding the 'Date' column from correlation calculation
correlation_matrix = df.drop(columns=['Date']).corr()
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
save_plot_to_pickle(heatmap, 'heatmap.pkl')  # Save the heatmap
plt.show()
```
